```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Task 2
df = pd.read_csv('train.csv')
df.sample(5)

```

```python
#WE are fillimg Null with Nan as 'No Category'
df['Product_Category_2']=df['Product_Category_3'].fillna(0).astype(int)
df['Product_Category_3']=df['Product_Category_3'].fillna(0).astype(int)
```

```python
print(df.isnull().sum())
```

```python
# Now We are looking How factcors affecting the Purchase 
#  Relationship [Gender vs Purchase]
plt.figure(figsize=(7,6))
sns.barplot(x='Gender', y='Purchase', data=df, hue='Gender', palette='magma', legend=False)
plt.title('Average Puchase Amount By Gender')
plt.show()
```

```python
# Relationship: Age Group vs Purchase
# Sorting age categories for better visualization
age_order=sorted(df['Age'].unique())
plt.figure(figsize=(9,6))
sns.boxplot(x= 'Age',y = 'Purchase',data =df, order=age_order)
plt.title('Purchase Distribution Across Age Groups')
plt.show()
```

```python
#  Relationship: Correlation Heatmap
# This shows how numerical variables relate to each other

sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap='RdYlGn')
plt.title('Correlation Matrix of Sales Data')
plt.show()
```
